```python
from PIL import Image
from numpy import array
im_mat = Image.open(r"C:/Users/user/Desktop/images.jpg")
im_ar = array(im_mat)
```

```python
im_ar
```

```python
def is_diag(matrix):
    for i in range(0, len(matrix[0])): 
        for j in range(0, len(matrix[0])): 
            if (((i != j) & (matrix[i][j] != 0))).all(): 
                return False
    return True
```

```python
is_diag(im_ar)
```

```python
def isuppertriangular(matrix):
    for i in range(1, len(matrix)):
        for j in range(0, i):
            if(matrix[i][j] != 0).all():
                    return False
    return True
```

```python
isuppertriangular(im_ar)
```
